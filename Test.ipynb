```python
x = data.drop('y', axis=1)
y = data['y']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x_train.info()
```

```python
x_train.describe()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
x_train_scaled[:5]
```

```python
x_test.head()
```

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
log_reg.fit(x_train_scaled, y_train)
y_pred=log_reg.predict(x_test_scaled)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
```

```python
from sklearn.svm import SVC
svm = SVC(kernel = 'linear')
svm.fit(x_train_scaled, y_train)
y_pred_svm=svm.predict(x_test_scaled)
```

```python
accuracy_svm = svm.score(x_test_scaled, y_test)
accuracy_svm
```

```python
svm.coef_
```

```python
from sklearn.metrics import confusion_matrix
y_test_array = np.array(y_test)
y_pred_array = np.array(y_pred_svm)

cm = confusion_matrix(y_test_array, y_pred_array)
sns.heatmap(cm, annot=True, fmt="d", cmap="Reds")
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV

dec_tree_params = [{'max_depth': list(range(2,10)), 'max_features': list(range(5,20)), 
                    'max_leaf_nodes': list(range(5, 10))}]
grid_search_cv = GridSearchCV(DecisionTreeClassifier(random_state=42), dec_tree_params)
grid_search_cv.fit(x_train_scaled, y_train)
```

```python
best_model=grid_search_cv.best_estimator_
```

```python
grid_search_cv.best_params_
```

```python
accuracy = best_model.score(x_test_scaled, y_test)
accuracy
```

```python
from sklearn.tree import export_graphviz

feature_names = data.columns.drop('y')
class_names = ['B', 'M']

export_graphviz( 
 grid_search_cv.best_estimator_,
 out_file=('g_dec_tree.dot'),
 feature_names=feature_names,
 class_names=class_names,
 filled=True,
)
```

```python
import graphviz 
graph = graphviz.Source.from_file('g_dec_tree.dot')
graph.format = 'png'
graph.render(filename='g_dec_tree', cleanup=True)
graph
```
